```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/lena256.jpg')
print(img.shape)
gimg1=cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
print(gimg1.shape)
#plt.imshow(gimg1, cmap='gray')
cv2_imshow(gimg1)
```

```python
def saturation(val):
  if val > 255:
    return 255
  elif val < 0:
    return 0
  else:
    return val
```

```python
kernel3 = np.ones((3,3),np.float32)/9
kernel5 = np.ones((5,5),np.float32)/25
kernel7 = np.ones((7,7),np.float32)/49
```

```python
output =cv2.filter2D(gimg1,-1,kernel5)
cv2_imshow(output)
```

```python
output =cv2.filter2D(gimg1,-1,kernel7)
cv2_imshow(output)
```

```python
gOut = np.zeros((gimg1.shape[0],gimg1.shape[1]), dtype=np.ubyte)
```

```python
for i in range(1,gimg1.shape[0]-1):
  for j in range(1,gimg1.shape[1]-1):
    sum = 0
    for k in range(-1,2):
      for h in range(-1,2):
        sum += gimg1[i+k, j+h] * kernel3[k+1,h+1]
      sum = saturation(sum)
      gOut[i,j] = sum

cv2_imshow(gOut)
```

```python
output =cv2.filter2D(output,-1,kernel3)
cv2_imshow(output)
```

```python
kernel3[0][0] = -1/9
kernel3[0][1] = -1/9
kernel3[0][2] = -1/9
kernel3[1][0] = -1/9
kernel3[1][1] = -1/9
kernel3[1][2] = -1/9
kernel3[2][0] = -1/9
kernel3[2][1] = -1/9
kernel3[2][2] = -1/9
```

```python
output =cv2.filter2D(gimg1,-1,kernel3)
#cv2_imshow(output)
output = output * 5
cv2_imshow(output)
```
